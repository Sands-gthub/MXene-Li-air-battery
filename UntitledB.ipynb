```python
import pandas as pd
df = pd.read_excel('KnownwithfeatureB.xlsx')
df
```

```python
redF = ['M1W', 'M1IE', 'M1R', 'M1MP', 'M1χ', 'M1CE', 'M1VS', 'M1VD', 'M1VF',
       'M2W', 'M2V', 'M2IE', 'M2R', 'M2MP', 'M2χ', 'M2CE', 'M2VS', 'M2VD',
       'M2VF', 'T1W', 'T1R', 'T1MP', 'T1VP', 'T1-M1', 'M1-X', 'X-M2', 'M2-T2',
       'SA', 'SW', 'M1XΔIE', 'M2XΔIE', 'M1XΔχ', 'M2XΔχ', 'M2TΔχ', 'B']
df0=df[redF]
df1 = df0[df0["B"]>-6.5]
df1
```

```python
feature_index=[]
for i in range(1,36):
  feature_index.append(i)

df2 = df1.copy()
df2.columns = feature_index

col_dict = dict(zip(df1.columns, df2.columns))

x = df2.drop([35], axis = 1)
y = df2[35]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)

from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
extr = ExtraTreesRegressor(bootstrap=False, ccp_alpha=0.0,
                    max_depth=12, max_leaf_nodes=None,
                    max_samples=None, min_impurity_decrease=0.0,
                    min_samples_leaf=4,
                    min_samples_split=5, min_weight_fraction_leaf=0.0,
                    n_estimators=100, n_jobs=None, oob_score=False,
                    random_state=42, verbose=0, warm_start=False)

extr.fit(x_train, y_train)
train_pred = extr.predict(x_train)
test_pred = extr.predict(x_test)

extr_MAE_train = mean_absolute_error(y_train, train_pred)
extr_MAE_test = mean_absolute_error(y_test, test_pred)
extr_r2_train = r2_score(y_train, train_pred)
extr_r2_test = r2_score(y_test, test_pred)

extr_MAE_train, extr_MAE_test, extr_r2_train, extr_r2_test
```

```python
import numpy as np
from sklearn.model_selection import RepeatedKFold,cross_val_score,KFold,LeaveOneOut
```

```python
STD = []
mae = []
kf = KFold(n_splits=10, shuffle = True, random_state=42)
scores = cross_val_score(extr, x, y, scoring='neg_mean_absolute_error', cv=kf)
pscore = -1*scores
sd = np.std(pscore)
mae.append(pscore)
STD.append(sd)
np.mean(mae), np.mean(STD)
```

```python
STD = []
mae = []
rs = []
for i in range(100):
  rs.append(i)
  kf = KFold(n_splits=10, shuffle = True, random_state=i)
  scores = cross_val_score(extr, x, y, scoring='neg_mean_absolute_error', cv=kf)
  pscore = -1*scores
  sd = np.std(pscore)
  mean_mae = np.mean(pscore)
  mean_std = np.mean(sd)
  mae.append(mean_mae)
  STD.append(mean_std)
#np.mean(mae), np.mean(STD), mae
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
mae = []
r2 = []
Num_features = []
for i in range(4,35,3):
  Num_features.append(i)
  k_features = i  # Number of top features to select using SIS
  selector = SelectKBest(score_func=f_regression, k=k_features)
  x_selected = selector.fit_transform(x, y)
  rkf = RepeatedKFold(n_splits=10, n_repeats=3, random_state=6)
  # Perform cross-validation for MAE and R2 scores
  mae_scores = cross_val_score(extr, x_selected, y, scoring='neg_mean_absolute_error', cv=rkf, n_jobs=None)
  mae_scores = -mae_scores
  mae.append(np.mean(mae_scores))
  r2_scores = cross_val_score(extr, x_selected, y, scoring='r2', cv=rkf, n_jobs=None)
  r2.append(np.mean(r2_scores))
```
