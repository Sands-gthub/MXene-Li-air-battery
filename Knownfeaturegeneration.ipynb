```python
import pandas as pd
KnownA = pd.read_excel("/content/A_energies.xlsx")
UnknownA = pd.read_excel("/content/Alldata_mostfeatures_A.xlsx")
```

```python
KnownA.columns
```

```python
KnownA['Names'] = KnownA['Name']
KnownA.columns
```

```python
UnknownA.columns
```

```python
UnknownA
```

```python
KnownA
```

```python
KnownwithfeatureA = UnknownA[UnknownA['Names'].isin(KnownA['Names'])]
KnownwithfeatureA
```

```python
KnownwithfeatureA.reset_index(drop=True, inplace=True)
KnownwithfeatureA
```

```python
KnownwithfeatureA['A'] = KnownA['A']
KnownwithfeatureA
```

```python
KnownwithfeatureA.to_excel('KnownwithfeatureA.xlsx', index = False)
```
