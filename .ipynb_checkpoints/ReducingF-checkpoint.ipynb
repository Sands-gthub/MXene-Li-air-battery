```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_excel('/content/KnownwithfeatureA.xlsx')
df.columns
```

### M1 features

```python
colM1 = ['M1W', 'M1AN', 'M1P', 'M1G', 'M1V', 'M1IE', 'M1R', 'M1MP', 'M1BP', 'M1χ','M1CE', 'M1GVE', 'M1VS', 'M1VP', 'M1VD', 'M1VF', 'M1USV', 'M1UPV', 'M1UDV', 'M1UFV', 'A']
df1 = df[colM1]
df1
```

```python
cor_plot=df1.corr()
plt.figure(figsize=(12, 10), dpi=600)

sns.heatmap(cor_plot, cmap='coolwarm', vmax=1.0, vmin=-1.0, linewidths=0.1, fmt='.2f', annot=True, annot_kws={"size": 5, "color":'black'}, square=False)
```

```python
colM1new = ['M1W', 'M1IE', 'M1R', 'M1MP', 'M1χ','M1CE', 'M1VS', 'M1VD', 'M1VF', 'A']
df1 = df[colM1new]
cor_plot=df1.corr()
plt.figure(figsize=(12, 10), dpi=600)

sns.heatmap(cor_plot, cmap='coolwarm', vmax=1.0, vmin=-1.0, linewidths=0.1, fmt='.2f', annot=True, annot_kws={"size": 5, "color":'black'}, square=False)
```
