```python
import pandas as pd
df = pd.read_excel("Reaction energies A.xlsx")
df
```

```python
# Define a function to extract atoms from the name after "relax_POSCAR_"
import re
def extract_atoms(name):
    match = re.search(r'relax_POSCAR_(.*)', name)
    if match:
        atoms_part = match.group(1)
        atoms = re.findall(r'[A-Z][a-z]*', atoms_part)
        return atoms
    else:
        return []
```

```python
df['Atoms'] = df['Name'].apply(extract_atoms)
df
```

```python
M1 = []
M2 = []
X = []
T1 = []
T2 = []
for i in df['Atoms']:
  if len(i) == 5:
    M1.append(i[0])
    M2.append(i[1])
    X.append(i[2])
    T1.append(i[3])

  elif len(i) == 7:
    M1.append(i[0])
    M2.append(i[1])
    X.append(i[2])
    T1.append(''.join(i[3:5]))

  elif len(i) == 9:
    M1.append(i[0])
    M2.append(i[1])
    X.append(i[2])
    T1.append(''.join(i[3:6]))
```

```python
df["M1"] = M1
df["M2"] = M2
df["X"] = X
df["T1"] = T1
df["T2"] = T1
```

```python
df
```

```python
df = df.drop(["Atoms"], axis=1)
df
```

```python
df.to_excel("A energies.xlsx",index=False)
```
