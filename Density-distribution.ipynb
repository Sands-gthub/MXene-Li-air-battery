```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_excel('/content/Predicted Reaction all density.xlsx')
df
```

```python
#For density plot

plt.rcParams["font.weight"]="bold"
plt.rcParams["axes.labelweight"]="bold"
plt.rcParams["figure.autolayout"] = True

sns.kdeplot(data=df[df['T1'] == 'Br']['Ediff'], color='#1f77b4', label='Br', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'Cl']['Ediff'], color='#ff7f0e', label='Cl', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'CN']['Ediff'], color='#2ca02c', label='CN', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'NCO']['Ediff'], color='#d62728', label='NCO', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'NCS']['Ediff'], color='#9467bd', label='NCS', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'O']['Ediff'], color='#8c564b', label='O', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'OCN']['Ediff'], color='#e377c2', label='OCN', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'PO']['Ediff'], color='#7f7f7f', label='PO', linewidth=2, fill=True)
sns.kdeplot(data=df[df['T1'] == 'SCN']['Ediff'], color='#bcbd22', label='SCN', linewidth=2, fill=True)

plt.xlabel('Ediff (eV)', fontsize=20)
plt.ylabel('Density', fontsize=20)
#plt.title('Density Plot', fontsize=18)
#plt.xlim(0, 2)
plt.xticks(range(-1, 7), fontsize=15)
plt.xlim(-1, 5)
plt.yticks(fontsize=15)
# Show the legend
plt.legend(loc="upper right", ncol=1, fontsize=12)
plt.savefig('density_T.png', dpi=700, bbox_inches="tight", transparent=True)
# Show the plot
plt.show()
```

```python
plt.rcParams["font.weight"]="bold"
plt.rcParams["axes.labelweight"]="bold"
plt.rcParams["figure.autolayout"] = True
plt.rcParams["figure.figsize"] = [6, 4]
Ediff = list(df['Ediff'])
T = list(df['T1'])

colors_map = {'Br':'#1f77b4', 'Cl':'#ff7f0e', 'CN':'#2ca02c', 'NCO':'#d62728', 'NCS':'#9467bd', 'O':'#8c564b', 'OCN':'#e377c2', 'PO':'#7f7f7f', 'SCN':'#bcbd22'}

unique_T = ['Br','Cl','CN','NCO','NCS','O','OCN','PO','SCN']
colors = [colors_map[t] for t in unique_T]

# Create the scatter plot
for i, t in enumerate(unique_T):
    indices = [idx for idx, val in enumerate(T) if val == t]
    plt.scatter(
        [T[idx] for idx in indices],
        [Ediff[idx] for idx in indices],
        c='white',
        label=t,
        facecolors='white',       # Remove fill in the center
        edgecolors=colors[i],    # Set the edge color
        linewidths=1.5             # Set the edge thickness
    )

plt.xlabel('Terminal Groups', fontsize=20)
plt.ylabel('Ediff (eV)', fontsize=20)
#plt.title('Density Plot of ML Predicted BE for C and N', fontsize=18)

plt.xticks(fontsize=15, rotation=90)
plt.yticks(range(0,6,1),fontsize=15)
plt.ylim(-0.5,5.5)
# Show the legend
#plt.legend(loc="upper left", ncol=2, fontsize=15)
plt.savefig('distribution_T.png', dpi=700, bbox_inches="tight", transparent=True)
# Show the plot
plt.show()
```
